```python
options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.maximize_window()


url = 'https://divar.ir/s/tehran/buy-residential/sohrevardi'
driver.get(url)

finally_links = []

# Scroll Function
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        
        try:
                
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            try:
                # test = driver.find_elements('//*[@id="post-list-container-id"]/div[2]/div/button')
                # print(test[0].text)
                button = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="post-list-container-id"]/div[2]/div/button')))
                button_text = button.text
                print(button_text)
                 # Find the button by its XPath (change the XPath as per your button's location)
                # button = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//button[@id='button_id']")))
                # Click the button
                button.click()
            except Exception as e:
                print(e)
                pass
            boxes = driver.find_elements(By.CLASS_NAME, "post-list__widget-col-c1444")

            links = []
            # finally_links = []

            for i in boxes:
                links.append(i.find_element(By.TAG_NAME, 'a'))
                
            for i in links:
                # print(i.get_attribute('href'))
                finally_links.append(i.get_attribute('href'))

            time.sleep(2)

            new_height = driver.execute_script("return document.body.scrollHeight")

            if new_height == last_height:
                break

            last_height = new_height
            
            
            
        except:
            pass

scroll_to_bottom(driver)
```
