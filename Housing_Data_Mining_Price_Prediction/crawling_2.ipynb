```python
# convert Persian digits to English digits
def persian_to_english(persian_str):
    persian_digits = '۰۱۲۳۴۵۶۷۸۹'
    english_digits = '0123456789'
    translation_table = str.maketrans(persian_digits, english_digits)
    return persian_str.translate(translation_table)

# translator=Translator()
# translator.translate('۴', src='fa', dest='en').text

a = []
b = []
c = []

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

for i in range(len(df)):
    url = df[i]
    # Open the website
    driver.get(f'{url}')

    # Find elements and append their text to the lists
    time.sleep(0.5)
    a_elements = driver.find_elements(By.CSS_SELECTOR, '#app > div:nth-child(1) > div > div > main > article > div > div.kt-col-5 > section:nth-child(1) > div.post-page__section--padded > div:nth-child(3) > div.kt-base-row__end.kt-unexpandable-row__value-box > p')
    time.sleep(0.5)
    b_elements = driver.find_elements(By.CSS_SELECTOR, '#app > div:nth-child(1) > div > div > main > article > div > div.kt-col-5 > section:nth-child(1) > div.post-page__section--padded > table:nth-child(1) > tbody > tr > td:nth-child(1)')
    time.sleep(0.5)
    c_elements = driver.find_elements(By.CSS_SELECTOR, '#app > div:nth-child(1) > div > div > main > article > div > div.kt-col-5 > section:nth-child(1) > div.post-page__section--padded > table:nth-child(1) > tbody > tr > td:nth-child(3)')
    time.sleep(0.5)
    
    # time.sleep(2)
    # a_elements = driver.find_elements(By.XPATH, '//*[@id="app"]/div[1]/div/div/main/article/div/div[1]/section[1]/div[4]/div[2]')
    # time.sleep(2)
    # b_elements = driver.find_elements(By.XPATH, '//*[@id="app"]/div[1]/div/div/main/article/div/div[1]/section[1]/div[4]/table[1]/tbody/tr/td[1]')
    # time.sleep(2)
    # c_elements = driver.find_elements(By.XPATH, '//*[@id="app"]/div[1]/div/div/main/article/div/div[1]/section[1]/div[4]/table[1]/tbody/tr/td[3]')
    # time.sleep(2)
    
    # Extract and clean the text for price
    price_text = a_elements[0].text if a_elements else "NULL"
    price_match = re.search(r'\d{1,3}(?:٬\d{3})*', price_text) if price_text != "NULL" else None
    price_value = int(price_match.group(0).replace('٬', '')) if price_match else "NULL"
    a.append(price_value)

    # Convert Persian digits to English digits
    meter_text = persian_to_english(b_elements[0].text) if b_elements else "NULL"
    room_text = persian_to_english(c_elements[0].text) if c_elements else "NULL"

    b.append(meter_text)
    c.append(room_text)

# print the lists with the cleaned values
print("Prices:", a)
print("Meters:", b)
print("Number of Rooms:", c)

# Close the driver
# driver.quit()
```

```python
# Create a DataFrame
data = {
    'Price': a,
    'Meters': b,
    'Number of Rooms': c
}

df_new = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df_new.to_csv('Final_Data.csv', index=False)

# Print the DataFrame to verify
print(df_new)
```
